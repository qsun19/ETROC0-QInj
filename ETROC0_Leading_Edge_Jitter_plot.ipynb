```python
import pandas as pd
#% matplotlib inline
from pandas import ExcelWriter
from pandas import ExcelFile

#df_highpwr_sim = pd.read_excel('leading_edge_jitter_high_pwr.xlsx',sheet_name='sim')
#df_highpwr_sim = pd.read_excel('leading_edge_jitter_high_pwr.xlsx',sheet_name='cdet-2p_c1-1p')
#df_highpwr_sim = pd.read_excel('leading_edge_jitter_high_pwr.xlsx',sheet_name='cdet-3.4p_c1-1p')
#df_highpwr_sim = pd.read_excel('leading_edge_jitter_high_pwr.xlsx',sheet_name='cdet-3.4p_c1-1p-1V')
#df_highpwr_sim = pd.read_excel('leading_edge_jitter_high_pwr.xlsx',sheet_name='cdet-3.4p_c1-1p-1VPA')
df_highpwr_sim = pd.read_excel('leading_edge_jitter_high_pwr.xlsx',
                               sheet_name='cdet-3.4p_c1-10f-1p2VPA') ## should be with the fast pulse
df_highpwr_sim
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as font_manager
font = font_manager.FontProperties(family='Arial', weight='normal',style='normal',size=8)

x_sim = df_highpwr_sim['tt_x']
x_sim = x_sim*1e15

lj_ff = df_highpwr_sim['ff_y']*1e12
lj_fs = df_highpwr_sim['fs_y']*1e12
lj_sf = df_highpwr_sim['sf_y']*1e12
lj_ss = df_highpwr_sim['ss_y']*1e12
lj_tt = df_highpwr_sim['tt_y']*1e12

x_mea = df_highpwr_sim['mea_x']
lj_mea = df_highpwr_sim['mea_y']


fig, ax = plt.subplots(dpi=600)
ax.plot(x_sim, lj_ff, marker = 's')
ax.plot(x_sim, lj_fs, marker = '^')
ax.plot(x_sim, lj_sf, marker = 'v')
ax.plot(x_sim, lj_ss, marker = '.')
ax.plot(x_sim, lj_tt, marker = ',')
ax.plot(x_mea, lj_mea, marker = 'o', color = '#000000')

ljLegend=('FF', 'FS', 'SF', 'SS', 'TT', 'Measured')

ax.set(xlabel='Injected Charge (fC)', ylabel='Leading Edge Jitter (ps)',
       title='Leading Edge Jitter at High Power')
ax.grid()
ax.legend(ljLegend, fancybox = True, shadow = False, ncol = 2)

ax.tick_params(direction='in')

ax.set_xlim([0, 35])
ax.set_ylim([0, 40])

#fig.savefig("test.png")
fig.savefig("leadingedgejitter_highpwr.pdf")
plt.show()
```
